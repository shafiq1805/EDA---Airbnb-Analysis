### What did you know about your dataset?

There are 10052 missing values in last_review and reviews_per_month. This suggests that listings with missing last reviews might also have missing values in the 'reviews_per_month' column.

Also there are 21 and 16 missing vales in host_name and name respectively. These missing values might be due to incomplete or optional information during the listing creation.

Hence, we can replace host_name and name with other and,   last_review and reviews_per_month with zeros.

````python
#Replacing Null values with 0.
null_columns=['last_review','reviews_per_month']
for col in null_columns:
  df1[col].fillna(0,inplace=True)

  # Replacing NA values with 'others'
dnull_columns=['host_name','name']
for col2 in dnull_columns:
  df1[col2].fillna("Others",inplace=True)

df1.isnull().sum().sort_values(ascending = False)
````

#### **Chart - 1 - Pricing Analysis**





````python
# Chart - 1 visualization code
plt.figure(figsize=(12, 8))
sns.boxplot(x='room_type', y='price', data=df1)
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
````

````python
sorted_df = df1.groupby('room_type')['price'].mean().reset_index()
sorted_df
````

##### 1. Why did you pick the specific chart?

Boxplots are commonly used in exploratory data analysis (EDA) because they provide a good balance between simplicity and informativeness. Also, The boxplot allows for easy comparison of price distributions across different room types. It provides a visual representation of the range and variability within each category.

##### 2. What is/are the insight(s) found from the chart?

1. According to the Price Distribution by Room Type chart, the most expensive room type on Airbnb is the Entire home/apt, followed by the Private room and Shared room.

2. One possible insight from this chart is that Airbnb users are willing to pay a premium for privacy and convenience.

* Entire home/apt listings offer the highest level of privacy and convenience, as guests have the entire space to themselves.

* Private room listings offer less privacy than Entire home/apt listings, but they are still more private than Shared room listings.

* Shared room listings are the least expensive option, but they also offer the least privacy, as guests share the space with other guests.

Overall, the Price Distribution by Room Type chart provides insights into the pricing and demand dynamics of the Airbnb market. It shows that Entire home/apt listings are the most expensive and in-demand room type, while Shared room listings are the least expensive and least in-demand room type.

#### **Chart - 5 - Seasonal Trends**

````python
# Chart - 5 visualization code
plt.figure(figsize=(12, 8))
sns.histplot(df1['availability_365'], bins=30, kde=True)
plt.title('Availability Trends Throughout the Year')
plt.xlabel('Availability (in days)')
plt.ylabel('Frequency')
plt.show()
````

##### 1. Why did you pick the specific chart?

A Histogram can  provide a clear and intuitive visual representation of the distribution of availability across different time periods. This visualization method makes it easy to observe peaks or troughs in availability, helping to identify busy and slow periods.

Histplots are easy to interpret and communicate, making them suitable for conveying seasonal trends to a diverse audience, including individuals who may not have a strong statistical background.

##### 2. What is/are the insight(s) found from the chart?

The analysis of the chart indicates a predominant trend among guests, with the majority opting for shorter stays, typically less than 10 days. Notably, around one-fifth of the total guests fall into the category of extended stays, surpassing 350 days. In contrast, a relatively smaller portion of guests chooses stays ranging from 10 days to 350 days. This distribution highlights a significant preference for shorter durations among the majority, while a notable segment of guests exhibits a preference for considerably longer stays, with fewer guests falling in the intermediate stay duration range.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Some of Impacts are:

**Short Stays Dominance:**:  Most Airbnb guests stay for less than 10 days. This means that Airbnb should focus on attracting short-term guests like offering discounts and promotions.

**Extended Stays Segment:** A significant number of Airbnb guests stay for more than 350 days. Airbnb could offer loyalty programs or discounts for long-term guests. It could also make it easier for long-term guests to find apartments that are suitable for their needs.

**Negative Growth Mitigation:** Airbnb could survey its guests to find out why they are not staying for longer periods of time. It could also experiment with different pricing and marketing strategies to attract more guests in this duration range.
